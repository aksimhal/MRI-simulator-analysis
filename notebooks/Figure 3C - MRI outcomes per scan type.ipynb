```python
import numpy as np 

%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
N = 7
scan_labels = ['Scan #1 (T1w)', 'Scan #2 (R-fMRI)', 'Scan #3 (T-fMRI)',\
               'Scan #4 (T-fMRI)', 'Scan #5 (R-fMRI)', 'Scan #6 (T2w)', 'Scan #7 (DTI)']
individual_pass = np.array([188, 173, 122, 116, 108,  84,  28])/201 * 100
individual_fail = np.array([13,    4,   2,   1,   0,  11,  28])/201 * 100
not_attempted =   np.array([0,    24,  77,  84,  93, 106, 145])/201 * 100

ind = np.arange(N)    # the x locations for the groups
width = 0.8       # the width of the bars: can also be len(x) sequence
f = plt.figure(figsize=(8, 6))
p1 = plt.bar(ind, individual_pass, width)
p2 = plt.bar(ind, individual_fail, width,
             bottom=individual_pass)
p3 = plt.bar(ind, not_attempted, width,
             bottom=individual_pass+individual_fail, color='gray')

plt.ylabel('Percentage of children (%)')
plt.title('MRI outcomes')
plt.xticks(ind, scan_labels, rotation='vertical')
plt.ylim([0, 130])
plt.legend((p1[0], p2[0], p3[0]), ('Pass', 'Fail', 'Not Completed'))
plt.rcParams.update({'font.size': 16})

plt.grid(False)

plt.tight_layout()
plt.show()
```
