```python
import numpy as np 
import pandas as pd
import seaborn as sns

%matplotlib notebook
import matplotlib.pyplot as plt

import matplotlib.ticker as mtick
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# fn = 'data.xlsx' #insert actual data here 
df = pd.read_excel(fn)
```

```python
df = df[df['MPRAGE'].isin(['pass', 'maybe', 'fail'])]
```

```python
df = df.reset_index()
```

```python
fd_list = [] 
label_list = [] 
for n in range(0, len(df)): 
    if df['REST1'][n] == 'pass' or df['REST1'][n] == 'maybe': 
        fd_list.append(df['rest1_median'][n])
        label_list.append('Pass')
    if df['REST1'][n] == 'fail': 
        fd_list.append(df['rest1_median'][n])
        label_list.append('Fail')

plt.figure()
histdf = pd.DataFrame({'motion':fd_list, 'Outcome':label_list})
sns.histplot(histdf, x="motion", hue='Outcome', bins=20, stat="probability", alpha=1)
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
plt.xlim([0, 0.4])
plt.ylim([0, 0.25])

plt.ylabel('Percentage of children (%)')
plt.xlabel('Median FD (mm)')
plt.legend(['Fail', 'Pass'])
plt.title('Median FD for R-fMRI 1')
plt.grid(False)
plt.rcParams.update({'font.size': 16})
plt.tight_layout()
```
