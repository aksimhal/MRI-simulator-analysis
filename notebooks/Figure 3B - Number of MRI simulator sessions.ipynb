```python
import numpy as np 

%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
fail_counts = np.array([15, 7, 3, 3, 1])/250 * 100
fail_visit_instance = [1, 2, 3, 4, 5]
```

```python
pass_counts = np.array([150, 41, 20, 8, 2])/250 * 100
pass_visit_instance = [1, 2, 3, 4, 5]
```

```python
plt.figure()
plt.bar(pass_visit_instance, pass_counts)
plt.bar(fail_visit_instance, fail_counts, color='C1', bottom=pass_counts)

plt.rcParams.update({'font.size': 16})
plt.xlabel('MRI simulator training sessions (counts)')
plt.ylabel('Percentage of children (%)')
plt.title('Number of MRI simulator sessions')
plt.legend(['Pass', 'Fail'])
plt.tight_layout()
```
